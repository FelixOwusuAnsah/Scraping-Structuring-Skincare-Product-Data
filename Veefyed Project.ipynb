```python
import pandas as pd
import requests as r
from bs4 import BeautifulSoup
import re
```

```python
homepage = r.get("https://qudobeauty.com/")
```

```python
soup = BeautifulSoup(homepage.text, 'html.parser')
```

```python
# Identifying the product links on the homepage
# Looping through all anchor tags to find product links
product_links = []

for a in soup.find_all("a", href=True):
    href = a["href"]
    if "/product/" in href:
        product_links.append(href)

product_links = list(set(product_links))
len(product_links)
```

```python
product_links
```

```python
def scrape_product(url):
    # Request individual product page
    response = r.get(url)
    def scrape_product(url):
        response = r.get(url)
    soup = BeautifulSoup(response.text, "html.parser")

    # Extract product name
    name_tag = soup.find("h1", class_="product_title")
    product_name = name_tag.get_text(strip=True) if name_tag else None

    # Extract main product image
    image_tag = soup.find("img", class_="wp-post-image")
    image_url = image_tag["src"] if image_tag else None

    # Return raw structure (some fields intentionally left empty for later cleaning)
    return {
        "product_name": product_name,
        "brand": None,
        "category": None,
        "ingredients": None,
        "size_packaging": None,
        "image_url": image_url,
        "product_page_url": url
    }
```

```python
# Having an overlook of the website data
products_data = []

for url in product_links:
    products_data.append(scrape_product(url))

# Convert scraped data into a DataFrame
df = pd.DataFrame(products_data)
df
```

```python
# downloading the dataset
df.to_csv("veefyed_day1_products.csv", index=False)
```

   # Data Normalisation
# After inspecting the initial dataset, I observed that several attributes, such as 
# brand, category, and size were embedded within the product_name field rather than 
# being explicitly structured. To improve consistency and usability of the dataset, 
# A post-processing normalisation step was applied.

```python
# Category keywords derived from inspection of product_name values
category_map = {
    "pad": "Pad",
    "serum": "Serum",
    "cream": "Cream",
    "mask": "Mask",
    "cleanser": "Cleanser",
    "cleansing oil": "Cleansing Oil",
    "oil": "Cleansing Oil",
    "toner": "Toner",
    "essence": "Essence",
    "capsule": "Cream",
    "shampoo": "Shampoo",
    "conditioner": "Conditioner"
}
```

```python
def normalize_from_name(row):
    name = row["product_name"]

    # Skip rows without valid names
    if not isinstance(name, str):
        return row

    # Separated the brand when included in the product name
    if " – " in name:
        brand, remainder = name.split(" – ", 1)
        row["brand"] = brand.strip()
        name = remainder

    # Extracted the size of information embedded in the product name
    size_match = re.search(r"(\d+\s?(ml|g|pcs|pc|pads))", name, re.IGNORECASE)
    if size_match:
        row["size_packaging"] = size_match.group(1)
        name = name.replace(size_match.group(1), "").strip()

    # Inferred category conservatively based on observed keywords
    for keyword, category in category_map.items():
        if keyword in name.lower():
            row["category"] = category
            break

    row["product_name"] = name.strip().rstrip(",")

    # Cleaned the final product name
    return row
```

```python
# Apply post-processing normalization to the dataset
df = df.apply(normalize_from_name, axis=1)
df
```

```python
# Standardize size formatting for consistency
def standardize_size(size):
    if not isinstance(size, str):
        return size
    size = size.lower().strip()
    size = size.replace("ml", " ml").replace("g", " g").replace("pcs", " pcs")
    return " ".join(size.split())

df["size_packaging"] = df["size_packaging"].apply(standardize_size)
```
